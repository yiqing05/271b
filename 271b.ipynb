```
import pandas as pd
data = pd.read_csv('checkpoint/mpii/res-50/log.txt',sep='\t')
# data2 = pd.read_csv('checkpoint/mpii/hg-1/log.txt',sep='\t')
```

```
train_acc=data['Train Acc']
val_acc=data['Val Acc']
train_loss = data['Train Loss']
val_loss = data['Val Loss']
print(len(train_acc))
```

```
# import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(range(1,81), train_acc,label='train acc')
plt.plot(range(1,81), val_acc,label='val acc')
plt.xlabel('iterations')
plt.ylabel('Accuracy')
plt.legend()
plt.subplot(122)
plt.plot(range(1,81), train_loss,label='train loss')
plt.plot(range(1,81), val_loss,label='val aloss')
plt.xlabel('iterations')
plt.ylabel('Loss')
plt.legend()
```
